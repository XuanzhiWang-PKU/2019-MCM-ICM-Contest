按年份切分

```python
alldata = []
alldata.append(excel_orig[excel_orig['YYYY'] == 2010])
alldata.append(excel_orig[excel_orig['YYYY'] == 2011])
alldata.append(excel_orig[excel_orig['YYYY'] == 2012])
alldata.append(excel_orig[excel_orig['YYYY'] == 2013])
alldata.append(excel_orig[excel_orig['YYYY'] == 2014])
alldata.append(excel_orig[excel_orig['YYYY'] == 2015])
alldata.append(excel_orig[excel_orig['YYYY'] == 2016])
alldata.append(excel_orig[excel_orig['YYYY'] == 2017])
```

```python
sum([len(d) for d in alldata])
```

所有县的代码

```python
all_counties = set(excel_orig['FIPS_Combined'].tolist())
len(all_counties)
```

```python
totalNum = 0
county2drugtotals = []
countyCode2Name = {}
for d in alldata:
    countytotal = {}
    for i in range(len(d)):
        data = d.iloc[i]
        countytotal[data['FIPS_Combined']] = data['TotalDrugReportsCounty']
        countyCode2Name[data['FIPS_Combined']] = data['State'] + ', ' + data['COUNTY']
    county2drugtotals.append(countytotal)
    totalNum += len(countytotal)
print('Total: {}'.format(totalNum))
```

```python
Name2Code = {v: k for k, v in countyCode2Name.items()}
```

```python
for countyTotal in county2drugtotals:
    for c in all_counties:
        if c not in countyTotal.keys():
            countyTotal[c] = 0
```

```python
locations = pd.read_excel('../2019_MCMProblemC_DATA/CountyLocations.xlsx')
```

```python
df2write = pd.DataFrame(columns=['Name', 'Code', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017'])
```

```python
def findLocation(name):
    for i in range(len(locations)):
        if name == locations.iloc[i]['state'] + ', ' + locations.iloc[i]['county']:
            return locations.iloc[i]['latitude'], locations.iloc[i]['longitude']
    return 0, 0
def find201x(countycode, index):
    return county2drugtotals[index].get(countycode, 0)
```

```python
idx = 0
for c in all_counties:
    data2insert = []
    data2insert.append(countyCode2Name[c])
    data2insert.append(c)
    data2insert.append(find201x(c, 0))
    data2insert.append(find201x(c, 1))
    data2insert.append(find201x(c, 2))
    data2insert.append(find201x(c, 3))
    data2insert.append(find201x(c, 4))
    data2insert.append(find201x(c, 5))
    data2insert.append(find201x(c, 6))
    data2insert.append(find201x(c, 7))
    df2write.loc[idx] = data2insert
    idx += 1
```
