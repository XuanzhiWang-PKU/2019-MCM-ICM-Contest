```python
import pandas as pd
import numpy as np
```

```python
excel_orig = pd.read_excel('../2019_MCMProblemC_DATA/MCM_NFLIS_Data.xlsx', sheet_name='Data')
```

```python
commonlyPrescribed = ['Hydromorphone', 'Hydrocodone', 'Oxycodone', 'Oxymorphone', 'Oxymorphone', 'Ethylmorphine', 'Buprenorphine']
heroin = ['Heroin']
all = {
    'Heroin': {
        
    },
    'Other Synthetic Opioids': {
        
    },
    'Commonly Prescribed Opioids':{
        
    }
}
for i in range(len(excel_orig)):
    data = excel_orig.iloc[i]
    countyName = data['State'] + ', ' + data['COUNTY']
    print(countyName)
    for k in all.keys():
        all[k][countyName] = [0, 0, 0, 0, 0, 0, 0, 0]
```

```python
for i in range(len(excel_orig)):
    data = excel_orig.iloc[i]
    subName = data['SubstanceName']
    countyName = data['State'] + ', ' + data['COUNTY']
    year = data['YYYY']
#     for k in all.keys():
#         if all[k].get(countyName) == None:
#             all[k].update(countyName = [0, 0, 0, 0, 0, 0, 0, 0])
    
    if subName in heroin:
        all['Heroin'][countyName][year - 2010] += 1
    elif subName in commonlyPrescribed:
        all['Commonly Prescribed Opioids'][countyName][year - 2010] += 1
    else:
        all['Other Synthetic Opioids'][countyName][year - 2010] += 1
```

```python
compontMatrix = pd.read_excel('CompontMatrix.xlsx')
compontMatrix = compontMatrix.values.T
print(compontMatrix.shape)
```

```python
for k in all.keys():
    df = pd.DataFrame(all[k]).T
    matData = df.sort_index(axis=0, ascending = True).values
    pd.DataFrame(np.dot(compontMatrix, matData)).to_excel(k + '.xlsx')
```
