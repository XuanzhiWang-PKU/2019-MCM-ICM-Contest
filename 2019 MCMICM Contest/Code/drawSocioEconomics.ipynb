```python
import pandas as pd
import numpy as np
import plotly.offline as py
import plotly.figure_factory as ff
```

```python
def readCsv(year):
    path = '../2019_MCMProblemC_DATA/ACS_' + str(year)[-2:] + '_5YR_DP02/ACS_' + str(year)[-2:] + '_5YR_DP02_with_ann.csv'
    return pd.read_csv(path).drop(0)
```

```python
data10 = readCsv(10)
```

```python
df = pd.DataFrame()
df['fips'] = data10['GEO.id2']
df['totalHouses'] = data10['HC01_VC03']
df['totalFamilies'] = data10['HC01_VC04']
df['totalSchools'] = data10['HC01_VC80']
df['ratioSingle'] = 1 - data10['HC03_VC44'].astype('float64') / 100
df['ratioM2F'] = data10['HC01_VC35'].astype('float64') / data10['HC01_VC42'].astype('float64')
```

```python
df
```

### 学校数量

```python
endpts = [100, 300, 500, 1000, 3000, 5000, 10000]
def drawData(fips, values, filename):
    greencolor = [
        'rgb(247,252,253)',
        'rgb(229,245,249)',
        'rgb(204,236,230)',
        'rgb(153,216,201)',
        'rgb(102,194,164)',
        'rgb(65,174,118)',
        'rgb(35,139,69)',
        'rgb(0,88,36)'
    ]

    fig = ff.create_choropleth(
        fips=fips, values=values,
        scope=['KY', 'OH', 'WV', 'VA', 'PA'],
        county_outline={'color': 'rgb(255,255,255)', 'width': 0.5},
        state_outline={'color': 'rgb(255,255,255)', 'width': 3},
        legend_title='Total Schools',
        show_state_data=True,
        show_hover=True,
        colorscale=greencolor,
        simplify_county=0,
        simplify_state=0,
        binning_endpoints = endpts,
    )
    fig['layout']['legend'].update({'x': 0})
    fig['layout']['annotations'][0].update({'x': -0.12, 'xanchor': 'left'})
    py.plot(fig, filename=filename)
fips = df['fips'].tolist()
values = df['totalSchools'].astype('float64').tolist()
drawData(fips, values, 'totalSchools.html')
```
